```python
df = pd.read_csv('fracture.csv')
df.head()
```

```python
bmi = df['weight_kg'] / (df['height_cm']/100) * (df['height_cm']/100)
df['bmi'] = bmi
```

```python
df
```

```python
a = df['sex']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
a = le.fit_transform(a)

df['sex'] = a
```

```python
df
```

```python
x = df[['age','sex','bmi','bmd']]
x
```

```python
y = df['fracture']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier

kc = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
model = kc.fit(x_train,y_train)

y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
plt.scatter(y_test,y_pred,marker=8,linewidths=12)
plt.title('Scatter Plot!')
plt.xlabel('Testing Data!')
plt.ylabel('Prediction Data!')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```
