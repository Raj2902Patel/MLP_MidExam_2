```python
df = pd.DataFrame({'Name':['A','B','C'],'Number':[10,20,30]})
df
```

```python
df.rename(columns={'Number':'Roll_Number'})
```

```python
df = pd.read_csv('fracture.csv')
df
```

```python
bmi = df['weight_kg'] / (df['height_cm']/100) * (df['height_cm']/100)
df['bmi'] = bmi
df
```

```python
a = df['sex']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
a = le.fit_transform(a)

df['sex'] = a
```

```python
df
```

```python
x = df[['age','sex','bmi','bmd']]
x
```

```python
y = df['fracture']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.svm import SVC

s = SVC(kernel='sigmoid')

model = s.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
plt.scatter(y_test,y_pred,marker=8,linewidths=12)
plt.title('Scatter Plot')
plt.xlabel('Testing Data')
plt.ylabel('Prediction Data')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```
